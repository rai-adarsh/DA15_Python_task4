```python
## exp1 : users = ['user1','user2','user3']

## check if username is having  is odd or even value in end

users=["user1","user2","user3","user4","user5","user6","user7","user8","user9","user10","user11","user12","user13","user14","user15"]

for user in users:
    print("user=",user)

    num = int(user[4])
    if num % 2 == 0:
        print(f"{user} ends with an even number.")
    else:
        print(f"{user} ends with an odd number.")
```

```python
## exp2: areaCode = [01302,02320,12323,13434,02323]

## check if area code start with 0 or something else 

area_code= ["01302","02320","12323","13434","02323","12236","02546","14578","14236","00231","04789","08974","13547"]

for code in area_code:
    print('area code =', code)
    if code.startswith("0"):
        print('starts with 0')
    else:
        print('starts with 1')
    print('--')
```

```python
## exp3: Checking if names start with a vowel

names= ["adarsh","sagar","osama","ugagaaa","prateek","nikita","abhishek","imli","sugar","aata","linledin"]

vowels = "a","e","i","o","u"
for name in names:
    print("name =", name)
    if name[0] in vowels:
        print("starts with a vowel")
    else:
        print("starts with another alphabet")
    print("--")

```

```python
## exp:4 Checking if words ends with letter 'a'

words=["akansha","abhishek","abhilasha","chetan","nikita","anjali","pranay","ajit","shraddha","osama","ayushi","chaitanya","saurabhi","anna"]

for word in words:
    print('words =', word)
    if word.endswith("a"):
        print("word ends with 'a' ")
    else:
        print("word doesn't ends with a")
    print('--')
```

```python
## exp5: Checking if passwords are alphanumeric

passwords = ["pass123", "hello@world", "secure99", "onlyletters", "123456","4789","qasw%987","78459","asdewq87","jaiho567","meriMarji","4579ADDy"]

for password in passwords:
    print("passwords=",password)

    if password.isalnum():
        print("password is alpha numeric")
    else:
        print("password is not alpha numeric")
    print("-----")

```

```python
## exp6: Checking if phone numbers contain only digits

p_no =["123456","99845","1456@as","16727894","14as546987#","9987456","6547812365","456fds35","4578912"]

for number in p_no:
    print("phone_number",number)

    if number.isnumeric():
        print("number is numeric")
    else:
        print("number is not numeric")

```

```python
## exp7: checking if all the g_name is in lower case

g_name=["Addy","adarsh","ZXC","asdf","cvbnmk","AASDWasds","atif","BhaiJaan","Huhhh","hmmmm","kHatarnaak"]

for name in g_name:
    print("name=",name)

    if name.islower():
        print("name is in lower case")
    else:
        print("name is not in lower case")
```

```python
## exp8: Converting a list of words to uppercase

words=["golgappe","biryani","chicken","eggs","tinde","icegola","namkeen","custard","pepsi","hajmola","hingoli","dahi"]

for word in words:
    print("Original_word=",word)

    print("Upper_Case_word=",word.upper())
    print("---")
```
